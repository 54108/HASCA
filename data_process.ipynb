```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"SimHei.ttf", size=14)

def show_data(filepath):
    data = np.loadtxt(filepath)
    data_to_show = data.flatten()
    counter = Counter(data_to_show)
    values = list(counter.keys())
    frequencies = list(counter.values())
    plt.plot(values, frequencies,marker=",", linestyle='-', color='blue')
    plt.title(filepath[-9:-4], fontproperties=font)
    plt.xlabel('数字', fontproperties=font)
    plt.ylabel('频率', fontproperties=font)
    plt.grid(True)
    plt.show()
    print(min(data_to_show), max(data_to_show))
```
